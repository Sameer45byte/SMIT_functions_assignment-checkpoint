## Write a function is_strong(password) that checks if the password:

* Has at least 8 characters

* Contains numbers

* Contains uppercase letters

* Contains lowercase letters

### Return True or False.

```python
def is_strong(password):
    if len(password) < 8:
        return False

    has_number = False
    has_upper = False
    has_lower = False

    for char in password:
        if char.isdigit():
            has_number = True
        elif char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True

    if has_number and has_upper and has_lower:
        return True
    else:
        return False
```

```python
print(is_strong("Hello123"))  
print(is_strong("hello123"))   
print(is_strong("HELLO123"))   
print(is_strong("Hello"))      
```

## Write a function process_transaction(balance, amount, transaction_type) that supports:

* "deposit" ‚Üí add amount

* "withdraw" ‚Üí subtract amount

* "transfer" ‚Üí subtract amount + 1% transfer fee

### Return the updated balance.
### If amount > balance (for withdraw/transfer), return "Transaction Failed".

```python
def process_transaction(balance, amount, transaction_type):
    
    if transaction_type == "deposit":
        balance += amount
        return balance
    
    elif transaction_type == "withdraw":
        if amount > balance:
            return "Transaction Failed"
        balance -= amount
        return balance
    
    elif transaction_type == "transfer":
        fee = amount * 0.01
        total_deduction = amount + fee
        
        if total_deduction > balance:
            return "Transaction Failed"
        balance -= total_deduction
        return balance
    
    else:
        return "Invalid Transaction Type"
```

```python
print(process_transaction(1000, 200, "deposit"))
print(process_transaction(1000, 200, "withdraw"))
print(process_transaction(1000, 200, "transfer"))
print(process_transaction(1000, 2000, "withdraw"))
```

## Write a function analyze_results(marks_list) 
### that returns:
    Highest marks

    Lowest marks
    
    Average
    
    Pass rate percentage
    Passing marks = 50
    
### Return the data in a dictionary.

```python
def analyze_results(marks_list):
    
    highest = max(marks_list)
    lowest = min(marks_list)
    average = sum(marks_list) / len(marks_list)
    
    # Passing marks = 50
    passed_students = 0
    for marks in marks_list:
        if marks >= 50:
            passed_students += 1
    
    pass_rate = (passed_students / len(marks_list)) * 100
    
    return {
        "Highest Marks": highest,
        "Lowest Marks": lowest,
        "Average": average,
        "Pass Rate (%)": pass_rate
    }
```

```python
marks = [45, 78, 90, 33, 67, 50]
result = analyze_results(marks)

print(result)
```

## Airline Seat Booking With Constraints

### Write a function book_seat(seats, seat_requested):

### Constraints:

    If seat_requested not in seats ‚Üí return "Invalid Seat"
    
    If seat is "booked" ‚Üí return "Seat Already Booked"
    
    Otherwise book the seat and return updated seats dictionary

### Example seats dictionary:
{
  "1A": "free",
  "1B": "booked",
  "2A": "free"
}

```python
def book_seat(seats, seat_requested):
    
    if seat_requested not in seats:
        return "Invalid Seat"
    
    if seats[seat_requested] == "booked":
        return "Seat Already Booked"
    
    seats[seat_requested] = "booked"
    return seats
```

```python
seats = {
    "1A": "free",
    "1B": "booked",
    "2A": "free"
}
print(book_seat(seats, "1A")) 
print(book_seat(seats, "1B"))
print(book_seat(seats, "3A"))
```

## Write a function calculate_tax(salary) using slabs:

* 0‚Äì50,000 ‚Üí 0%

* 50,001‚Äì100,000 ‚Üí 5%

* 100,001‚Äì200,000 ‚Üí 12%

* Above 200,000 ‚Üí 20%

### Return total tax payable.

```python
def calculate_tax(salary):
    
    tax = 0
    
    if salary <= 50000:
        tax = 0
    
    elif salary <= 100000:
        tax = (salary - 50000) * 0.05
    
    elif salary <= 200000:
        tax = (50000 * 0.05) + (salary - 100000) * 0.12
    
    else:
        tax = (50000 * 0.05) + (100000 * 0.12) + (salary - 200000) * 0.20
    
    return tax
```

```python
print(calculate_tax(40000))
print(calculate_tax(75000))
print(calculate_tax(150000))
print(calculate_tax(300000))
```

## Create a Python program that:

### Defines a function called generate_salary(employee)

####    The employee parameter will be a dictionary containing:

        "name" ‚Üí employee name
        
        "basic" ‚Üí basic salary
        
        "overtime_hours" ‚Üí extra hours worked
        
        "performance" ‚Üí string ("excellent", "good", "average", "poor")

### The function must:

#####    ‚úî 1. Calculate Gross Salary
    
    Gross = basic salary
    
    overtime pay
    
    performance bonus
    
#####    ‚úî 2. Overtime Pay Rules
    
    1 overtime hour = 300 PKR
    
####    ‚úî 3. Performance Bonus Rules
    
    excellent ‚Üí 20% of basic
    
    good ‚Üí 10%
    
    average ‚Üí 5%
    
    poor ‚Üí 0%
    
#####    ‚úî 4. Apply Deductions
    
    Income tax = 12% of gross
    
    Medical insurance = 1000 PKR
    
#####    ‚úî 5. Return & print a full salary slip with this format:
        Employee: Ali
        Basic Salary: 50000
        Overtime Pay: 6000
        Performance Bonus: 10000
        Gross Salary: 66000
        Income Tax (12%): 7920
        Medical Insurance: 1000
        Net Salary: 57080

### üìù Example Input:
    employee = {
        "name": "Ali",
        "basic": 50000,
        "overtime_hours": 20,
        "performance": "excellent"
    }

### üìù Example Output:
    Employee: Ali
    Basic Salary: 50000
    Overtime Pay: 6000
    Performance Bonus: 10000
    Gross Salary: 66000
    Income Tax (12%): 7920
    Medical Insurance: 1000
    Net Salary: 57080

```python
def generate_salary(employee):
    
    name = employee["name"]
    basic = employee["basic"]
    overtime_hours = employee["overtime_hours"]
    performance = employee["performance"].lower()
    
    overtime_pay = overtime_hours * 300
    
    if performance == "excellent":
        bonus = basic * 0.20
    elif performance == "good":
        bonus = basic * 0.10
    elif performance == "average":
        bonus = basic * 0.05
    else:
        bonus = 0
    
    gross = basic + overtime_pay + bonus
    
    income_tax = gross * 0.12
    medical_insurance = 1000
    
    net_salary = gross - income_tax - medical_insurance
    
    print(f"Employee: {name}")
    print(f"Basic Salary: {basic}")
    print(f"Overtime Pay: {overtime_pay}")
    print(f"Performance Bonus: {bonus}")
    print(f"Gross Salary: {gross}")
    print(f"Income Tax (12%): {income_tax}")
    print(f"Medical Insurance: {medical_insurance}")
    print(f"Net Salary: {net_salary}")
    
    return net_salary

employee = {
    "name": "Ali",
    "basic": 50000,
    "overtime_hours": 20,
    "performance": "excellent"
}

generate_salary(employee)
```

## Create a Python program that:

### Defines a function calculate_patient_bill(patient)

    patient is a dictionary with keys: "name", "days_admitted", "room_type", "tests"

### Rules:

    Room charges per day:
    
        "General": 2000
        
        "Semi-Private": 4000
        
        "Private": 6000
        
        Test charges = sum of all test costs in tests list
        
        Apply 10% insurance discount on total
        
        Print the full bill with patient name, room charges, test charges, discount, and net bill

### üìù Example Input:
    patient = {
        "name": "Ali",
        "days_admitted": 3,
        "room_type": "Private",
        "tests": [1500, 2000, 500]
    }

### üìù Example Output:
    Patient: Ali
    Room Charges: 18000
    Test Charges: 4000
    Subtotal: 22000
    Insurance Discount (10%): 2200
    Net Bill: 19800

```python
def calculate_patient_bill(patient):
    
    name = patient["name"]
    days = patient["days_admitted"]
    room_type = patient["room_type"]
    tests = patient["tests"]
    
    room_prices = {
        "General": 2000,
        "Semi-Private": 4000,
        "Private": 6000
    }
    
    room_charge_per_day = room_prices.get(room_type, 0)
    room_charges = room_charge_per_day * days
    
    test_charges = sum(tests)
  
    subtotal = room_charges + test_charges
    
    discount = subtotal * 0.10
    
    net_bill = subtotal - discount

    print(f"Patient: {name}")
    print(f"Room Charges: {room_charges}")
    print(f"Test Charges: {test_charges}")
    print(f"Subtotal: {subtotal}")
    print(f"Insurance Discount (10%): {discount}")
    print(f"Net Bill: {net_bill}")
    
    return net_bill

patient = {
    "name": "Ali",
    "days_admitted": 3,
    "room_type": "Private",
    "tests": [1500, 2000, 500]
}

calculate_patient_bill(patient)
```

## Create a Python program that:

### Defines a function course_progress(student)

    student is a dictionary: "name", "lessons_completed", "total_lessons", "quizzes" (list of scores)

### The function should:

    Calculate lesson completion %
    
    Calculate average quiz score
    
    Determine progress status:
    
    ‚â• 80% completion ‚Üí "On Track"
    
    50‚Äì79% ‚Üí "Needs Improvement"
    
    <50% ‚Üí "Behind"
    
    Print a progress report

### üìù Example Input:
    student = {
        "name": "Ubaid",
        "lessons_completed": 8,
        "total_lessons": 10,
        "quizzes": [80, 75, 90]
    }

### üìù Example Output:
    Student: Ubaid
    Lesson Completion: 80%
    Average Quiz Score: 81.7
    Progress Status: On Track

```python
def course_progress(student):
    
    name = student["name"]
    lessons_completed = student["lessons_completed"]
    total_lessons = student["total_lessons"]
    quizzes = student["quizzes"]
    
    completion_percentage = (lessons_completed / total_lessons) * 100
    
    average_quiz = sum(quizzes) / len(quizzes) if quizzes else 0
    
    if completion_percentage >= 80:
        status = "On Track"
    elif 50 <= completion_percentage < 80:
        status = "Needs Improvement"
    else:
        status = "Behind"
    
    print(f"Student: {name}")
    print(f"Lesson Completion: {completion_percentage:.0f}%")
    print(f"Average Quiz Score: {average_quiz:.1f}")
    print(f"Progress Status: {status}")
    
    return {
        "completion_percentage": completion_percentage,
        "average_quiz_score": average_quiz,
        "status": status
    }

student = {
    "name": "Ubaid",
    "lessons_completed": 8,
    "total_lessons": 10,
    "quizzes": [80, 75, 90]
}

course_progress(student)
```

## A hotel calculates room charges:

### Function room_charges(nights, room_type) ‚Üí rates:

    Standard = $100/night
    
    Deluxe = $200/night
    
    Suite = $400/night
    
    Function additional_services(services) ‚Üí Breakfast=$20, Spa=$50, Gym=$10
    
    Function apply_membership_discount(amount, membership_years) ‚Üí 5% discount if membership > 2 years, 10% if >5 years
    
    Function total_bill(nights, room_type, services, membership_years) ‚Üí prints detailed bill

### üì• Input Example:
    total_bill(4, "Deluxe", ["Breakfast", "Spa"], 6)

### üìù Example Output:
Room Charges: 800
Additional Services: 70
Membership Discount: 87
Total Amount: 783

```python
def room_charges(nights, room_type):
    rates = {
        "standard": 100,
        "deluxe": 200,
        "suite": 400
    }
    return nights * rates.get(room_type.lower(), 0)


def additional_services(services):
    service_cost = {
        "Breakfast": 20,
        "Spa": 50,
        "Gym": 10
    }
    return sum(service_cost.get(service, 0) for service in services)


def apply_membership_discount(amount, membership_years):
    if membership_years > 5:
        return amount * 0.10
    elif membership_years > 2:
        return amount * 0.05
    return 0


def total_bill(nights, room_type, services, membership_years):

    room_cost = room_charges(nights, room_type)
    service_cost = additional_services(services)

    subtotal = room_cost + service_cost
    discount = apply_membership_discount(subtotal, membership_years)

    total = subtotal - discount

    print("Room Charges:", room_cost)
    print("Additional Services:", service_cost)
    print("Membership Discount:", int(discount))
    print("Total Amount:", int(total))

total_bill(4, "Deluxe", ["Breakfast", "Spa"], 6)
```

## A school calculates student performance:

### Each student has marks in Math, Science, English, History

    Function average_score(marks) ‚Üí calculates average
    
    Function grade(avg) ‚Üí nested function assigns grade:
    
    ‚â•90 ‚Üí A, 80‚Äì89 ‚Üí B, 70‚Äì79 ‚Üí C, 60‚Äì69 ‚Üí D, <60 ‚Üí F
    
    Function honor_roll(avg, extracurricular) ‚Üí gives "Honor Roll" if avg ‚â•85 and extracurricular ‚â•3
    
    Function student_summary(student) ‚Üí prints name, average, grade, honor roll status

### üì• Input Example:
    student = {"name":"Ali", "marks":[85,90,80,95], "extracurricular":3}
    student_summary(student)

### üìù Example Output:
    Name: Ali
    Average Score: 87.5
    Grade: B
    Honor Roll: Yes

```python
def average_score(marks):
    return sum(marks) / len(marks)


def grade(avg):
    def assign_grade(score):
        if score >= 90:
            return "A"
        elif score >= 80:
            return "B"
        elif score >= 70:
            return "C"
        elif score >= 60:
            return "D"
        else:
            return "F"

    return assign_grade(avg)


def honor_roll(avg, extracurricular):
    if avg >= 85 and extracurricular >= 3:
        return "Yes"
    return "No"


def student_summary(student):

    name = student["name"]
    marks = student["marks"]
    extracurricular = student["extracurricular"]

    avg = average_score(marks)
    g = grade(avg)
    honor = honor_roll(avg, extracurricular)

    print("Name:", name)
    print("Average Score:", round(avg, 1))
    print("Grade:", g)
    print("Honor Roll:", honor)

student = {"name": "Ali", "marks": [85, 90, 80, 95], "extracurricular": 3}
student_summary(student)
```

## A store applies layered discounts:

    Function base_discount(price) ‚Üí 5%
    
    Function category_discount(price, category) ‚Üí Electronics=10%, Clothing=8%, Grocery=3%
    
    Function coupon_discoun
    
    Function final_price(price, category, coupon) ‚Üí nested calls calculate total discounts + final price

### üì• Input Example:
    final_price(500, "Electronics", "SAVE10")

### üìù Example Output:
    Price after base discount: 475
    Price after category discount: 427.5
    Price after coupon: 417.5
    Final Price: 451.3 (with 8% tax)

```python
def base_discount(price):
    return price * 0.05


def category_discount(price, category):
    category = category.lower()

    discounts = {
        "electronics": 0.10,
        "clothing": 0.08,
        "grocery": 0.03
    }

    return price * discounts.get(category, 0)


def coupon_discount(price, coupon):

    if coupon == "SAVE10":
        return 10
    elif coupon == "SAVE20" and price > 200:
        return 20
    return 0


def final_price(price, category, coupon):

    base = base_discount(price)
    price1 = price - base
    print("Price after base discount:", round(price1, 1))

    cat_disc = category_discount(price1, category)
    price2 = price1 - cat_disc
    print("Price after category discount:", round(price2, 1))

    coup_disc = coupon_discount(price2, coupon)
    price3 = price2 - coup_disc
    print("Price after coupon:", round(price3, 1))

    final_amount = price3 + (price3 * 0.08)

    print("Final Price:", round(final_amount, 1))

final_price(500, "Electronics", "SAVE10")
```
